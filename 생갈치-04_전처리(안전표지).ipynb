```python
# 가져다 쓸 모듈들
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 예쁜 matplotlib
import folium # 지도 시각화
import json # geojson 파일용
import geopandas as gpd # geojson 파일용
```

```python
filename = "30.대전광역시_차량등록현황_격자.geojson"
file = open(filename,encoding="utf8")
df2 = gpd.read_file(file)
df2
```

```python
filename = "3월22일제발마지막-4.json"
file = open(filename,encoding="utf8")
df = gpd.read_file(file)
df.drop(df.loc[df['index']==-1].index, inplace=True)
df.drop(df.loc[df['sign_met_cd']=='0'].index, inplace=True)
df=df.sort_values(by='index')
df=df.reset_index()
del df['level_0']
df
```

```python
df3=df2.copy()
df3['안전표지판 수']=0
df3['표지판-정주식']=0
df3['표지판-내민식']=0
df3['표지판-부착식']=0
df3['표지판-문형식']=0
df3['표지판종류-주의']=0
df3['표지판종류-규제']=0
df3['표지판종류-지시']=0
df3['표지판종류-보조']=0
df3
```

```python
for i in range (len(df['index'])):
    df3['안전표지판 수'][df['index'][i]]+=1
    if df['sign_met_cd'][i]=='1':
        df3['표지판-정주식'][df['index'][i]]+=1
    elif df['sign_met_cd'][i]=='2':
        df3['표지판-내민식'][df['index'][i]]+=1
    elif df['sign_met_cd'][i]=='3':
        df3['표지판-부착식'][df['index'][i]]+=1
    elif df['sign_met_cd'][i]=='4':
        df3['표지판-문형식'][df['index'][i]]+=1
    if df['sign_knd_cd_f'][i]=='1':
        df3['표지판종류-주의'][df['index'][i]]+=1
    elif df['sign_knd_cd_f'][i]=='2':
        df3['표지판종류-규제'][df['index'][i]]+=1
    elif df['sign_knd_cd_f'][i]=='3':
        df3['표지판종류-지시'][df['index'][i]]+=1
    elif df['sign_knd_cd_f'][i]=='4':
        df3['표지판종류-보조'][df['index'][i]]+=1
```
