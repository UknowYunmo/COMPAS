```python
# 가져다 쓸 모듈들
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 예쁜 matplotlib
import folium # 지도 시각화
import json # geojson 파일용
import geopandas as gpd # geojson 파일용
```

```python
filename = "12.대전광역시_인구정보(총인구).geojson"
file = open(filename,encoding="utf8")
df90 = gpd.read_file(file)
df90.rename(columns = {'val' : '인구정보(총인구)'}, inplace = True)
df90['인구정보(총인구)']=df90['인구정보(총인구)'].fillna(0)
df90
```

```python
filename = "13.대전광역시_인구정보(고령).geojson"
file = open(filename,encoding="utf8")
df91 = gpd.read_file(file)
df91.rename(columns = {'val' : '인구정보(고령)'}, inplace = True)
df91['인구정보(고령)']=df91['인구정보(고령)'].fillna(0)
df91
```

```python
filename = "14.대전광역시_인구정보(생산가능).geojson"
file = open(filename,encoding="utf8")
df92 = gpd.read_file(file)
df92.rename(columns = {'val' : '인구정보(생산가능)'}, inplace = True)
df92['인구정보(생산가능)']=df92['인구정보(생산가능)'].fillna(0)
df92
```

```python
filename = "15.대전광역시_인구정보(유소년).geojson"
file = open(filename,encoding="utf8")
df93 = gpd.read_file(file)
df93.rename(columns = {'val' : '인구정보(유소년)'}, inplace = True)
df93['인구정보(유소년)']=df93['인구정보(유소년)'].fillna(0)
df93
```

```python
temp_1=pd.merge(df90,df91)
temp_1
```

```python
temp_2=pd.merge(temp_1,df92)
temp_2
```

```python
df=pd.merge(temp_2,df93)
df
```
