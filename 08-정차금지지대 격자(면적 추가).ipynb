```python
# 가져다 쓸 모듈들
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 예쁜 matplotlib
import folium # 지도 시각화
import json # geojson 파일용
import geopandas as gpd # geojson 파일용
```

```python
filename = "08-정차금지지대 격자.json"
file = open(filename,encoding="utf8")
df90 = gpd.read_file(file)
```

```python
df91=df90.copy()
df91['정차금지지대 면적']=0.0
df91=df91.reset_index()
df91
```

```python
filename = "8.대전광역시_정차금지지대.geojson"
file = open(filename,encoding="utf8")
df = gpd.read_file(file)
df
```

```python
converted_json = df.to_json()

m = folium.Map(
    location=[36.34156,127.39704],
    zoom_start=11
)

folium.GeoJson(
    converted_json,
).add_to(m)

m
```

```python
t=gpd.overlay(df91,df,how='intersection')
t=t.reset_index()
t
```

```python
del t['level_0']
```

```python
t
```

```python
len(t['index'].to_list())
```

```python
len(t.area.to_list())
```

```python
for a,b in zip(t['index'].to_list(),t.area.to_list()):
    df91['정차금지지대 면적'][a]+=b
```

```python
df91['정차금지지대 면적'].sum()
```

```python
big=df91.sort_values(by='정차금지지대 면적',ascending=False).head(10)
big
```

```python
small=df91[df91['정차금지지대 수']>0].sort_values(by='정차금지지대 면적',ascending=True).head(10)
small
```

```python
converted_json = df.to_json()

m = folium.Map(
    location=[36.34156,127.39704],
    zoom_start=11
)

folium.GeoJson(
    converted_json,
).add_to(m)

converted_json = big.to_json()

folium.GeoJson(
    converted_json,
).add_to(m)

m
```

```python
converted_json = df.to_json()

m = folium.Map(
    location=[36.34156,127.39704],
    zoom_start=11
)

folium.GeoJson(
    converted_json,
).add_to(m)

converted_json = small.to_json()

folium.GeoJson(
    converted_json,
).add_to(m)

m
```
