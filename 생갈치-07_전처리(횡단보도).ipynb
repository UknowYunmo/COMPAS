### 2. 횡단보도 면적 격자화

```python
filename = "격자_07-횡단보도-1.json"
file = open(filename,encoding="utf8")
df90 = gpd.read_file(file)
df90
```

```python
df91=df90.copy()
df91['횡단보도 면적']=0.0
df91=df91.reset_index()
df91
```

```python
filename = "6.대전광역시_횡단보도.geojson"
file = open(filename,encoding="utf8")
df = gpd.read_file(file)
df
```

```python
t=gpd.overlay(df91,df,how='intersection')
t=t.reset_index()
t
```

```python
converted_json = df.to_json()

m = folium.Map(
    location=[36.34156,127.39704],
    zoom_start=11
)

folium.GeoJson(
    converted_json,
).add_to(m)

m
```
