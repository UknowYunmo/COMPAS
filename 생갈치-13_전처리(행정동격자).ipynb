```python
# 가져다 쓸 모듈들
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 예쁜 matplotlib
import folium # 지도 시각화
import json # geojson 파일용
import geopandas as gpd # geojson 파일용
```

```python
filename = "30.대전광역시_차량등록현황_격자.geojson"
file = open(filename,encoding="utf8")
df = gpd.read_file(file)
df
```

```python
filename = "27.대전광역시_행정경계(읍면동).geojson"
file = open(filename,encoding="utf8")
df11 = gpd.read_file(file)
df11
```

```python
converted_json = df11.to_json()

m = folium.Map(
    location=[36.34156,127.39704],
    zoom_start=11
)

folium.GeoJson(
    converted_json,
).add_to(m)

m
```

```python
sum(df11['geometry'].is_valid==False)
```

```python
# 유효하지 않은 행정구역 geometry가 있어 버퍼를 씌움
df12=df11.copy()
df12['geometry']=df11['geometry'].buffer(0)
```

```python
sum(df12['geometry'].is_valid==False)
```

```python
df['행정동']='미정'
```

```python
df
```

```python
index=[]
dong=[]
for i in range (len(df['geometry'])):
    print(i)
    for j in range (len(df12['geometry'])):
        if df['geometry'][i].within(df12['geometry'][j])==True:
            #print(index)
            #print(df12['ADM_DR_NM'][j])
            index.append(i)
            dong.append(df12['ADM_DR_NM'][j])
for a, b in zip(index,dong):
    df['행정동'][a]=b
```

```python
for a, b in zip(index,dong):
    df['행정동'][a]=b
```

```python
certain=df[df['행정동']!='미정']
certain
```

```python
converted_json = certain.to_json()

m = folium.Map(
    location=[36.34156,127.39704],
    zoom_start=11
)

folium.GeoJson(
    converted_json,
).add_to(m)

m
```

```python
rest=df[df['행정동']=='미정']
rest=rest.reset_index()
rest
```

```python
md=gpd.overlay(rest,df11,how='intersection')
md
```

```python
converted_json = md.to_json()

m = folium.Map(
    location=[36.34156,127.39704],
    zoom_start=11
)

folium.GeoJson(
    converted_json,
).add_to(m)

m
```

```python
md['면적']=md['geometry'].area
```

```python
md.sort_values(by='gid')
```

```python
md.groupby(by='gid').max()
```

```python
f=md.groupby(by='gid').max()
f['행정동']=f['ADM_DR_NM']
f
```

```python
index2=f['index'].to_list()
dong2=f['행정동'].to_list()
```

```python
for a, b in zip(index2,dong2):
    df['행정동'][a]=b
```

```python
df[df['행정동']!='미정']
```

```python
df.to_file("격자_13-행정별격자.json", driver="GeoJSON")
```
