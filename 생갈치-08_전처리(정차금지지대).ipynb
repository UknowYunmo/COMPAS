### 1. 정차금지지대 수 격자화

```python
# 가져다 쓸 모듈들
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 예쁜 matplotlib
import folium # 지도 시각화
import json # geojson 파일용
import geopandas as gpd # geojson 파일용
```

```python
filename = "30.대전광역시_차량등록현황_격자.geojson"
file = open(filename,encoding="utf8")
df90 = gpd.read_file(file)
df90
```

```python
df91=df90.copy()
df91
```

```python
df91['정차금지지대 수']=0
df91
```

```python
filename = "8.대전광역시_정차금지지대.geojson"
file = open(filename,encoding="utf8")
df = gpd.read_file(file)
df
```

```python
print(len(df91))
for i in range (len(df91['geometry'])):
    print(i)
    for j in range (len(df['geometry'])):
        if df91['geometry'][i].intersects(df['geometry'][j])==True:
            df91['정차금지지대 수'][i]+=1
```

```python
df91.to_file("격자_08-정차금지지대-1.json", driver="GeoJSON")
```

### 2. 정차금지지대 면적 격자화

```python
filename = "격자_08-정차금지지대-1.json"
file = open(filename,encoding="utf8")
df90 = gpd.read_file(file)
df90
```

```python
df91=df90.copy()
df91['정차금지지대 면적']=0.0
df91=df91.reset_index()
df91
```

```python
filename = "8.대전광역시_정차금지지대.geojson"
file = open(filename,encoding="utf8")
df = gpd.read_file(file)
df
```

```python
t=gpd.overlay(df91,df,how='intersection')
t=t.reset_index()
t
```

```python
converted_json = df.to_json()

m = folium.Map(
    location=[36.34156,127.39704],
    zoom_start=11
)

folium.GeoJson(
    converted_json,
).add_to(m)

m
```

```python
converted_json = t.to_json()

m = folium.Map(
    location=[36.34156,127.39704],
    zoom_start=11
)

folium.GeoJson(
    converted_json,
).add_to(m)

m
```
