### 1. 정차금지지대 수 격자화

```python
# 가져다 쓸 모듈들
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 예쁜 matplotlib
import folium # 지도 시각화
import json # geojson 파일용
import geopandas as gpd # geojson 파일용
```

```python
filename = "30.대전광역시_차량등록현황_격자.geojson"
file = open(filename,encoding="utf8")
df90 = gpd.read_file(file)
df90
```

```python
df91=df90.copy()
df91
```

```python
df91['중앙분리대 비율']=0
df91
```

```python
filename = "31.대전시_중앙분리대.geojson"
file = open(filename,encoding="utf8")
df = gpd.read_file(file)
df
```

```python
print(len(df91))
for i in range (len(df91['geometry'])):
    print(i)
    for j in range (len(df['geometry'])):
        if df91['geometry'][i].intersects(df['geometry'][j])==True:
            df91['중앙분리대 수'][i]+=1
df91
```

```python
df91.to_file("격자_11-중앙분리대-1", driver="GeoJSON")
```

### 2. 정차금지지대 면적 격자화

```python
filename = "격자_11-중앙분리대-1.json"
file = open(filename,encoding="utf8")
df90 = gpd.read_file(file)
df90['중앙분리대 면적']=0.0
df90=df90.reset_index()
df90
```

```python
df91=df90.copy()
df91
```

```python
filename = "31.대전시_중앙분리대.geojson"
file = open(filename,encoding="utf8")
df = gpd.read_file(file)
df
```

```python
t=gpd.overlay(df90,df,how='intersection')
t=t.reset_index()
t
```

```python
for a,b in zip(t['index'].to_list(),t.area.to_list()):
    df91['중앙분리대 면적'][a]+=b
```

```python
df91[df91['중앙분리대 수']>0]
```

```python
df91.to_file("격자_11-중앙분리대-2", driver="GeoJSON")
```
